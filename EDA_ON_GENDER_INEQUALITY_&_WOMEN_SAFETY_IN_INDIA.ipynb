**Data Cleaning** and standardizing the 'State' column by stripping spaces and converting to lowercase (this is used for convert lowercase to upper but upper and proper space already done)

```python
df["State"] =df["State"].str.strip().str.lower()
df['State'].unique()
```

```python
df.columns
```

**Extracting Total Crimes Column**

```python
df["Total crimes"] = df[['Rape', 'Kidnapping and abdiction', 'Dowry Deaths',
       'Assault on women', 'Insult to Modesty of Women', 'Domestic Violence',
       'Women Trafficking']].fillna(0).sum(axis=1)
```

```python
df
```

**Trends in Total Crimes Against Women in India (2001-2021)**

Objective:Â¶
To analyze and visualize the changes in the total number of crimes against women in India over the years from 2001 to 2021
To identify any significant increases or decreases in crime rates


The plot shows a general increase in crimes against women from 2001 to 2021, with noticeable fluctuations and a significant rise in recent years.

Increasing Trend: The National Crime Records Bureau (NCRB) reported a continuous rise in crimes against women, with incidents of rape increasing from approximately 18,000 cases in 2001 to over 32,000 in 2021.

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=df['Year'],y=df['Total crimes'])
plt.title("Trends in Total Crimes Against Women in India (2001-2021)")
plt.xlabel("Year")
plt.ylabel("Total Crimes")
plt.grid(True)
plt.savefig("Trends in Total Crimes Against Women in India (2001-2021).png")
plt.show()
```

**Top 10 States with the Highest Total Crimes Against Women in India (2001-2021)**

```python
statewise_crimes = df.groupby('State') ["Total crimes"].sum().reset_index().sort_values(by="Total crimes",ascending=False).reset_index(drop=True).head(10)
statewise_crimes
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="State", y="Total crimes", data=statewise_crimes, palette="Reds_d")
plt.title("Top 10 states with Highest crime rates")
plt.xlabel("State")
plt.ylabel("Total Crimes")
plt.xticks(rotation=90)
plt.savefig("Top 10 states with Highest crime rates.png")
plt.show()
```

**Comparison of Crimes Against Women**

```python
statewise_crimetypesum = df.groupby('State')[['Total crimes','Rape', 'Kidnapping and abdiction', 'Dowry Deaths',
                                'Assault on women', 'Insult to Modesty of Women',
                                'Domestic Violence', 'Women Trafficking']].sum().reset_index().sort_values(by="Total crimes",ascending=False).reset_index(drop=True).head(10)
statewise_crimetypesum
```

```python
melted = statewise_crimetypesum.melt(id_vars='State',
                                        value_vars=['Rape', 'Kidnapping and abdiction', 'Dowry Deaths',
                                                    'Assault on women', 'Insult to Modesty of Women',
                                                    'Domestic Violence', 'Women Trafficking'],
                                        var_name='Crime Type',
                                        value_name='Number of Crimes')
melted
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='State', y='Number of Crimes', hue='Crime Type', data=melted, palette='Paired')
plt.title('Crime Categories Across Top 10 States')
plt.xlabel('States')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45)
plt.show()
```

**Crime Type Distribution in India (2021)**

```python
year = 2021
df_year = df[df["Year"] == year]
crime_data2021 =df_year[['Rape', 'Kidnapping and abdiction', 'Dowry Deaths',
                        'Assault on women', 'Insult to Modesty of Women',
                        'Domestic Violence', 'Women Trafficking']].sum()
crime_data2021
```

**Crime Count in Maharashtra**

```python
mh = df[df["State"] == "maharashtra"]
mh_crime_counts = mh.iloc[:, 2:9].sum()
mh_crime_counts
```

```python
plt.figure(figsize=(10, 6))
plt.bar(mh_crime_counts.index, mh_crime_counts.values)
plt.title("Crime Count in Maharashtra")
plt.xlabel("Crime Type")
plt.ylabel("Number of cases")
plt.xticks(rotation=45)
plt.show()
```

**Piechart for data 2021**

```python
plt.figure(figsize=(10, 6))
plt.pie(crime_data2021, labels=crime_data2021.index, autopct="%1.1f%%", startangle= 90, colors=sns.color_palette("Set2"))
plt.title("Crime Type Distribution in India (2021)")
plt.tight_layout()
plt.show()
```

#### To visually represent the distribution of different crime types against women in India for the year 2021.

#### Domestic violence and assault against women were the most reported crimes in 2021, making up over 60% of the total.

#### women trafficking was the least reported crime in 2021, making up only 0.3% of the total.

Domestic violence:
### Domestic violence is a significant issue, especially during the COVID-19 lockdowns,when calls to helplines increased.
### For example, the National Commission for Women (NCW) received over 1,400 complaints of domestic violence in just the first month of the lockdown in 2020,
Dowry Deaths:
### Although dowry deaths represent a smaller percentage in the pie chart, they remain a significant issue in rural areas and specific states like Bihar and Uttar Pradesh.


```python
#Heatmap
correlation_matrix = df[['Rape', 'Kidnapping and abdiction', 'Dowry Deaths',
                        'Assault on women', 'Insult to Modesty of Women',
                        'Domestic Violence', 'Women Trafficking']].corr()
correlation_matrix
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='RdPu')
plt.title('Correlation Matrix of Crimes Against Women (2001-2021)')
plt.show()
```
